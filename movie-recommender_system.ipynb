```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("movies.csv")
credits = pd.read_csv("credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

#### Merging Both The Datasets

```python
#Merging on the basis of title
movies = movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
movies.head(1)
```

#### Removing the Unwanted Colums

```python
### WANTED COLUMNS - 
#genres
#id
#keywords
#title
#overview
#cast
#crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.info()
```

```python
movies.head(2) #The Main Data
```

```python
#Making tags 
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
#checking duplicate data
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
#We want to take out the name tags (Action, fantasy etc) and put them in another list. This is our preprocessing.
#Now this list is a string. we have to convert the string of list to list
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert) #Convert func gets applies
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
def get_cast(obj):
    M = []
    for i in ast.literal_eval(obj):
            M.append((i['name']))
    return M[0:3]
```

```python
movies["cast"] = movies["cast"].apply(get_cast)
```

```python
ast.literal_eval(movies['crew'][0])
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

```python
#converting overview to a list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
#movies.head()
```

```python
#We have to remove spaces from all the values (eg - Hello There - HelloThere)
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
#converting the tags list to string
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#converting to lowercase
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#Now we have to do text-vectorisation to calculate the similiarity between the strings
#We will convert the whole text to vectors
#We will use the bag of words method
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() #converting to numpy array
```

```python
vectors
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
#WE WILL APPLY STEMING TO MAKE THE DATA MORE POLISHED
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray() #converting to numpy array
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
#checking the similarity of words
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity #gets the similiarity score of movies with other movies
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#first we find the index of the movie
#then we will get similiarityscore list of that movie with other movies
#then we sort the list and get the first 5 movies

#first 5 movies - 
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('The Outsiders') #we get 5 movies
```
